```python
import sys
import os
from pyspark.sql.functions import *
from pyspark.sql.types import *

## Replace the workspace path with your own
sys.path.append(f"/Workspace/Users/thiruvengadamk16@gmail.com/Retail-And-Ecommerce-Analytics-Platform")
from src.paths import RAW_PRODUCTS_PATH, BRONZE_PRODUCTS_PATH
from src.utils import add_ingestion_timestamp
```

### Reading the Raw Products Files

```python
prod_df = spark.read.format("csv")\
    .option("header", "true")\
    .option("inferSchema", "true")\
    .load(RAW_PRODUCTS_PATH)
```

```python
display(prod_df.limit(5))
```

### Adding metadata columns ("ingestion_ts", "_source_file")

```python
prod_df = add_ingestion_timestamp(prod_df)
```

### Creating Bronze Products Table

```python
(
    prod_df
    .write
    .format("delta")
    .option("delta.columnMapping.mode", "name")
    .mode("overwrite")
    .saveAsTable(BRONZE_PRODUCTS_PATH)
)
```
