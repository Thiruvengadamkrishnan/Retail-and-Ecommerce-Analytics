```sql
SELECT 
    p.product_id,
    p.category,
    p.sub_category,
    t.total_quantity_sold,
    t.total_revenue
FROM ecommerce_catalog.ecommerce_gold1.top_selling_skus t
JOIN ecommerce_catalog.ecommerce_gold1.dim_product p
  ON t.product_id = p.product_id
ORDER BY t.total_revenue DESC
LIMIT 20;
```
